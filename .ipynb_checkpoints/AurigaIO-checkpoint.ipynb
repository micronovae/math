```python
# def get_data(basePATH,time,parttype,field):
#     res = ag.snapshot.loadSubset(basePATH,time,
#                                  parttype,field)
#     res = res[field]
#     return res

def mas_cent(pos,mas):
    
    res = [0,0,0]
    for i in range(len(pos)):
        res += pos[i]*mas[i]
    res = res/sum(mas)
    
    return res

def cent(pos):
    
    x = np.mean(pos[:,0])
    print(x)
    y = np.mean(pos[:,1])
    z = np.mean(pos[:,2])
    res = np.array([x,y,z])
    
    return res


def inertia_tensor(pos,mas,cent):
    
    '''
    pos = [[x0,y0,z0],[x1,y2,z2]...]
    mas = [m1,m2.....]
    cent = [x,y,z]
    '''
    tensor = np.zeros((3,3))
    j = 0 
    for i in range(len(mas)):
    
        tensor += mas[i]*((pos[i]-cent)*np.array([[pos[i][0]-cent[0]],
                                    [pos[i][1]-cent[1]],
                                    [pos[i][2]-cent[2]]]))
       
    return tensor

def angFor2Vector(v,x):
    '''
    input : vector v, vector x
    output : the angle from v to x
    '''
    v = np.array(v)
    x = np.array(x)
    v = v/np.sqrt(sum(v*v))
    x = x/np.sqrt(sum(x*x))

    theta = math.acos(sum(x*v))
    
    return theta

def CrossMat3D(v):
    
    v = np.array(v)
    if len(v) != 3:
        print('Dim error')
        
    mat = np.zeros((3,3))
    
    mat[0][1] = -v[2]
    mat[1][0] = v[2]
    mat[0][2] = v[1]
    mat[2][0] = -v[1]
    mat[1][2] = -v[0]
    mat[2][1] = v[0]
    
    return mat
        
    
def normalVector(v,x):
    
    '''v cross x'''
    v = np.array(v)
    x = np.array(x)
    v = CrossMat3D(v)
    k = np.dot(v,x)
    
    return k
    

def SO3Mat(v,x):
    '''
        from v to x
    '''
    theta = angFor2Vector(v,x)
    
    if (theta%2*np.pi<0.0001):
        res = np.identity(3)
        
    else:
        k = normalVector(v,x)
        k = k/np.sqrt(sum(k*k))
        K = CrossMat3D(k)
        res = (np.identity(3)+math.sin(theta)*K
               +(1-math.cos(theta))*np.dot(K,K))
    
    return res
        
        
```
